```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
df = pd.read_json("../data/combined_objects.json")

# If position and velocity are dictionaries, convert them into separate columns
if 'position' in df.columns:
    position_columns = df['position'].apply(pd.Series)
    df = pd.concat([df.drop(['position'], axis=1), position_columns], axis=1)

if 'velocity' in df.columns:
    velocity_columns = df['velocity'].apply(pd.Series)
    df = pd.concat([df.drop(['velocity'], axis=1), velocity_columns], axis=1)

# Fill NaNs for velocity columns with zero and convert to numeric
df[['vx', 'vy', 'vz']] = df[['vx', 'vy', 'vz']].fillna(0).apply(pd.to_numeric, errors='coerce')

# Simulate trajectories
days_to_simulate = 365 * 100000
seconds_in_day = 86400
simulation_seconds = days_to_simulate * seconds_in_day

df['x'] += df['vx'] * simulation_seconds
df['y'] += df['vy'] * simulation_seconds
df['z'] += df['vz'] * simulation_seconds

# Calculate distances to Earth (assuming Earth is at [0,0,0] for simplicity)
df['distance_to_earth'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)

# Determine collision risk
collision_threshold_km = 500000  # 100,000 km threshold
df['collision_risk'] = df['distance_to_earth'] < collision_threshold_km

# Visualize the data
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['x'], df['y'], df['z'], c='blue', label='Asteroids')
ax.scatter([0], [0], [0], color='red', marker='x', s=100, label='Earth')
ax.set_xlabel('X Position (km)')
ax.set_ylabel('Y Position (km)')
ax.set_zlabel('Z Position (km)')
ax.set_title('Asteroid Trajectories and Collision Risk')
ax.legend()
plt.show()

# Output asteroids with potential collision risk
print("Asteroids with Potential Collision Risk:")
print(df[df['collision_risk']])
```

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from ipywidgets import interact

# Load the asteroid data
df = pd.read_json("../data/asteroids.json")

# Expand position and velocity into separate columns
df[['x', 'y', 'z']] = df['position'].apply(lambda pos: pd.Series(pos))
df[['vx', 'vy', 'vz']] = df['velocity'].apply(lambda vel: pd.Series(vel))

# Define Earth's position
earth_position = np.array([149600000.0, 0.0, 0.0])  # in kilometers

# Simulation function with velocity adjustment
def simulate_and_plot(years_to_simulate, velocity_multiplier):
    simulation_seconds = years_to_simulate * 365 * 86400

    # Adjust velocities to move towards Earth with the given multiplier
    for i, row in df.iterrows():
        position = np.array([row['x'], row['y'], row['z']])
        direction_to_earth = earth_position - position
        distance_to_earth = np.linalg.norm(direction_to_earth)
        if distance_to_earth > 0:
            normalized_direction = direction_to_earth / distance_to_earth
            # Update velocities to move faster towards Earth
            df.at[i, 'vx'] = normalized_direction[0] * velocity_multiplier * np.log1p(distance_to_earth)
            df.at[i, 'vy'] = normalized_direction[1] * velocity_multiplier * np.log1p(distance_to_earth)
            df.at[i, 'vz'] = normalized_direction[2] * velocity_multiplier * np.log1p(distance_to_earth)

    # Update positions based on new velocities
    df['x'] += df['vx'] * simulation_seconds
    df['y'] += df['vy'] * simulation_seconds
    df['z'] += df['vz'] * simulation_seconds

    # Plot the results
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df['x'], df['y'], df['z'], color='blue', s=1, label='Asteroids')
    ax.scatter(*earth_position, color='green', s=500, label='Earth')

    ax.set_xlabel('X (km)')
    ax.set_ylabel('Y (km)')
    ax.set_zlabel('Z (km)')
    ax.set_title(f'Asteroids Trajectories after {years_to_simulate} years')
    ax.legend()
    plt.show()

# Interactive widgets
years_slider = widgets.IntSlider(min=100, max=10000, step=100, value=1000, description='Years:')
velocity_slider = widgets.FloatSlider(min=0.1, max=10.0, step=0.1, value=1.0, description='Velocity Multiplier:')

interact(simulate_and_plot, years_to_simulate=years_slider, velocity_multiplier=velocity_slider)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets

# Load data
df = pd.read_json("../data/asteroids.json")

# Extract position and velocity into separate columns
df['x'], df['y'], df['z'] = zip(*df['position'].apply(lambda pos: (pos['x'], pos['y'], pos['z'])))
df['vx'], df['vy'], df['vz'] = zip(*df['velocity'].apply(lambda vel: (vel['vx'], vel['vy'], vel['vz'])))

# Constants for Earth's position in km
earth_x, earth_y, earth_z = 149600000.0, 0.0, 0.0

# Function to update and visualize asteroid positions
def update_positions(years):
    seconds_in_year = 365.25 * 24 * 3600  # Account for leap year
    # Update positions based on velocities and years
    df['x_new'] = df['x'] + df['vx'] * years * seconds_in_year / 1000  # convert to km
    df['y_new'] = df['y'] + df['vy'] * years * seconds_in_year / 1000
    df['z_new'] = df['z'] + df['vz'] * years * seconds_in_year / 1000

    # Calculate distance to Earth
    df['distance_to_earth'] = np.sqrt((df['x_new'] - earth_x)**2 + 
                                      (df['y_new'] - earth_y)**2 + 
                                      (df['z_new'] - earth_z)**2)

    # Collision threshold (in km), set to a more reasonable range
    collision_threshold_km = 1000000  # For closer proximity collisions

    # Detect collisions
    collisions = df[df['distance_to_earth'] <= collision_threshold_km]

    # Visualization
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df['x_new'], df['y_new'], df['z_new'], color='blue', alpha=0.6, s=1, label='Asteroids')
    ax.scatter([earth_x], [earth_y], [earth_z], color='red', s=100, label='Earth')

    if not collisions.empty:
        ax.scatter(collisions['x_new'], collisions['y_new'], collisions['z_new'], 
                   color='orange', s=20, label='Collisions Detected')
        print("Collisions detected:")
        print(collisions[['id', 'distance_to_earth']])
    else:
        print("No collisions detected.")

    ax.set_xlabel('X (km)')
    ax.set_ylabel('Y (km)')
    ax.set_zlabel('Z (km)')
    ax.set_title(f'Simulation over {years} years')
    ax.legend()
    plt.show()

# Widget to adjust the number of years
year_slider = widgets.IntSlider(min=5, max=10000, step=50, value=1000, description='Years:')
widgets.interactive(update_positions, years=year_slider)
```
