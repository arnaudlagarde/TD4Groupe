```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
df = pd.read_json("../data/combined_objects.json")

# If position and velocity are dictionaries, convert them into separate columns
if 'position' in df.columns:
    position_columns = df['position'].apply(pd.Series)
    df = pd.concat([df.drop(['position'], axis=1), position_columns], axis=1)

if 'velocity' in df.columns:
    velocity_columns = df['velocity'].apply(pd.Series)
    df = pd.concat([df.drop(['velocity'], axis=1), velocity_columns], axis=1)

# Fill NaNs for velocity columns with zero and convert to numeric
df[['vx', 'vy', 'vz']] = df[['vx', 'vy', 'vz']].fillna(0).apply(pd.to_numeric, errors='coerce')

# Simulate trajectories
days_to_simulate = 365 * 100000
seconds_in_day = 86400
simulation_seconds = days_to_simulate * seconds_in_day

df['x'] += df['vx'] * simulation_seconds
df['y'] += df['vy'] * simulation_seconds
df['z'] += df['vz'] * simulation_seconds

# Calculate distances to Earth (assuming Earth is at [0,0,0] for simplicity)
df['distance_to_earth'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)

# Determine collision risk
collision_threshold_km = 500000  # 100,000 km threshold
df['collision_risk'] = df['distance_to_earth'] < collision_threshold_km

# Visualize the data
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['x'], df['y'], df['z'], c='blue', label='Asteroids')
ax.scatter([0], [0], [0], color='red', marker='x', s=100, label='Earth')
ax.set_xlabel('X Position (km)')
ax.set_ylabel('Y Position (km)')
ax.set_zlabel('Z Position (km)')
ax.set_title('Asteroid Trajectories and Collision Risk')
ax.legend()
plt.show()

# Output asteroids with potential collision risk
print("Asteroids with Potential Collision Risk:")
print(df[df['collision_risk']])
```

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from ipywidgets import interact

# Load the asteroid data
df = pd.read_json("../data/asteroids.json")

# Expand position and velocity into separate columns
df[['x', 'y', 'z']] = df['position'].apply(lambda pos: pd.Series(pos))
df[['vx', 'vy', 'vz']] = df['velocity'].apply(lambda vel: pd.Series(vel))

# Define Earth's position
earth_position = np.array([149600000.0, 0.0, 0.0])  # in kilometers

# Simulation function with velocity adjustment
def simulate_and_plot(years_to_simulate, velocity_multiplier):
    simulation_seconds = years_to_simulate * 365 * 86400

    # Adjust velocities to move towards Earth with the given multiplier
    for i, row in df.iterrows():
        position = np.array([row['x'], row['y'], row['z']])
        direction_to_earth = earth_position - position
        distance_to_earth = np.linalg.norm(direction_to_earth)
        if distance_to_earth > 0:
            normalized_direction = direction_to_earth / distance_to_earth
            # Update velocities to move faster towards Earth
            df.at[i, 'vx'] = normalized_direction[0] * velocity_multiplier * np.log1p(distance_to_earth)
            df.at[i, 'vy'] = normalized_direction[1] * velocity_multiplier * np.log1p(distance_to_earth)
            df.at[i, 'vz'] = normalized_direction[2] * velocity_multiplier * np.log1p(distance_to_earth)

    # Update positions based on new velocities
    df['x'] += df['vx'] * simulation_seconds
    df['y'] += df['vy'] * simulation_seconds
    df['z'] += df['vz'] * simulation_seconds

    # Plot the results
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df['x'], df['y'], df['z'], color='blue', s=1, label='Asteroids')
    ax.scatter(*earth_position, color='green', s=500, label='Earth')

    ax.set_xlabel('X (km)')
    ax.set_ylabel('Y (km)')
    ax.set_zlabel('Z (km)')
    ax.set_title(f'Asteroids Trajectories after {years_to_simulate} years')
    ax.legend()
    plt.show()

# Interactive widgets
years_slider = widgets.IntSlider(min=100, max=10000, step=100, value=1000, description='Years:')
velocity_slider = widgets.FloatSlider(min=0.1, max=10.0, step=0.1, value=1.0, description='Velocity Multiplier:')

interact(simulate_and_plot, years_to_simulate=years_slider, velocity_multiplier=velocity_slider)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets

# Load data
df = pd.read_json("../data/asteroids.json")

# Extract position and velocity into separate columns
df['x'], df['y'], df['z'] = zip(*df['position'].apply(lambda pos: (pos['x'], pos['y'], pos['z'])))
df['vx'], df['vy'], df['vz'] = zip(*df['velocity'].apply(lambda vel: (vel['vx'], vel['vy'], vel['vz'])))

# Constants for Earth's position in km
earth_x, earth_y, earth_z = 149600000.0, 0.0, 0.0

# Function to update and visualize asteroid positions
def update_positions(years):
    seconds_in_year = 365.25 * 24 * 3600  # Account for leap year
    # Update positions based on velocities and years
    df['x_new'] = df['x'] + df['vx'] * years * seconds_in_year / 1000  # convert to km
    df['y_new'] = df['y'] + df['vy'] * years * seconds_in_year / 1000
    df['z_new'] = df['z'] + df['vz'] * years * seconds_in_year / 1000

    # Calculate distance to Earth
    df['distance_to_earth'] = np.sqrt((df['x_new'] - earth_x)**2 + 
                                      (df['y_new'] - earth_y)**2 + 
                                      (df['z_new'] - earth_z)**2)

    # Collision threshold (in km), set to a more reasonable range
    collision_threshold_km = 1000000  # For closer proximity collisions

    # Detect collisions
    collisions = df[df['distance_to_earth'] <= collision_threshold_km]

    # Visualization
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df['x_new'], df['y_new'], df['z_new'], color='blue', alpha=0.6, s=1, label='Asteroids')
    ax.scatter([earth_x], [earth_y], [earth_z], color='red', s=100, label='Earth')

    if not collisions.empty:
        ax.scatter(collisions['x_new'], collisions['y_new'], collisions['z_new'], 
                   color='orange', s=20, label='Collisions Detected')
        print("Collisions detected:")
        print(collisions[['id', 'distance_to_earth']])
    else:
        print("No collisions detected.")

    ax.set_xlabel('X (km)')
    ax.set_ylabel('Y (km)')
    ax.set_zlabel('Z (km)')
    ax.set_title(f'Simulation over {years} years')
    ax.legend()
    plt.show()

# Widget to adjust the number of years
year_slider = widgets.IntSlider(min=5, max=10000, step=50, value=1000, description='Years:')
widgets.interactive(update_positions, years=year_slider)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
import json

# Load data
with open('../data/asteroids.json', 'r') as file:
    data = json.load(file)

# Convert JSON data to DataFrame
df = pd.DataFrame(data)

# Extract position and velocity
df['x'] = df['position'].map(lambda d: d['x'])
df['y'] = df['position'].map(lambda d: d['y'])
df['z'] = df['position'].map(lambda d: d['z'])
df['vx'] = df['velocity'].map(lambda d: d['vx'])
df['vy'] = df['velocity'].map(lambda d: d['vy'])
df['vz'] = df['velocity'].map(lambda d: d['vz'])

# Constants
earth_x, earth_y, earth_z = 149600000.0, 0.0, 0.0
seconds_in_year = 365.25 * 24 * 3600
collision_threshold_km = 50000  # 50,000 km for collision risk

# Time points to visualize (in years)
time_points = [0, 100, 200, 300]

fig = plt.figure(figsize=(15, 12))

for i, year in enumerate(time_points):
    ax = fig.add_subplot(2, 2, i + 1, projection='3d')
    # Calculate new positions
    x_new = df['x'] + df['vx'] * year * seconds_in_year / 1000
    y_new = df['y'] + df['vy'] * year * seconds_in_year / 1000
    z_new = df['z'] + df['vz'] * year * seconds_in_year / 1000

    # Distance to Earth
    distance_to_earth = np.sqrt((x_new - earth_x)**2 + (y_new - earth_y)**2 + (z_new - earth_z)**2)
    close_asteroids = distance_to_earth < collision_threshold_km

    # Plotting
    ax.scatter(x_new[~close_asteroids], y_new[~close_asteroids], z_new[~close_asteroids], color='black', s=1, alpha=0.5, label='Safe Asteroids')
    ax.scatter(x_new[close_asteroids], y_new[close_asteroids], z_new[close_asteroids], color='red', s=20, label='Collision Risk')
    ax.scatter([earth_x], [earth_y], [earth_z], color='blue', s=200, edgecolor='black', label='Earth')  # Make Earth more visible

    # Zoom in around Earth
    zoom_range = 100000  # Adjust to ensure Earth is visible
    ax.set_xlim(earth_x - zoom_range, earth_x + zoom_range)
    ax.set_ylim(earth_y - zoom_range, earth_y + zoom_range)
    ax.set_zlim(earth_z - zoom_range, earth_z + zoom_range)

    ax.set_xlabel('X (km)')
    ax.set_ylabel('Y (km)')
    ax.set_zlabel('Z (km)')
    ax.legend()
    ax.set_title(f"Asteroids relative to Earth at {year} years")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import json

# Load data
with open('../data/asteroids.json', 'r') as file:
    data = json.load(file)

# Convert JSON data to DataFrame
df = pd.DataFrame(data)

# Extract position and velocity
df['x'] = df['position'].map(lambda d: d['x'])
df['y'] = df['position'].map(lambda d: d['y'])
df['z'] = df['position'].map(lambda d: d['z'])
df['vx'] = df['velocity'].map(lambda d: d['vx'])
df['vy'] = df['velocity'].map(lambda d: d['vy'])
df['vz'] = df['velocity'].map(lambda d: d['vz'])

# Constants
earth_x, earth_y, earth_z = 149600000.0, 0.0, 0.0
seconds_in_year = 365.25 * 24 * 3600

# Time points to check (in years)
time_points = np.arange(0, 301, 10)

# Initialize list to store closest approach details
closest_approaches = []

# Iterate over each asteroid
for index, asteroid in df.iterrows():
    min_distance = float('inf')
    closest_year = 0
    closest_position = (0, 0, 0)
    
    # Check the distance to Earth at each time point
    for year in time_points:
        # Calculate new positions
        x_new = asteroid['x'] + asteroid['vx'] * year * seconds_in_year / 1000
        y_new = asteroid['y'] + asteroid['vy'] * year * seconds_in_year / 1000
        z_new = asteroid['z'] + asteroid['vz'] * year * seconds_in_year / 1000

        # Calculate distance to Earth
        distance_to_earth = np.sqrt((x_new - earth_x)**2 + (y_new - earth_y)**2 + (z_new - earth_z)**2)

        # Update if this is the closest approach
        if distance_to_earth < min_distance:
            min_distance = distance_to_earth
            closest_year = year
            closest_position = (x_new, y_new, z_new)

    # If the closest distance is within the collision threshold, store the details
    if min_distance < 30000:  # 30,000 km threshold for collision risk
        closest_approaches.append({
            'ID': asteroid['id'],
            'Closest Distance (km)': round(min_distance, 2),
            'Year of Closest Approach': closest_year,
            'Closest Position (x, y, z) km': tuple(round(pos, 2) for pos in closest_position),
            'Velocity (km/s)': tuple(round(v / 1000, 2) for v in (asteroid['vx'], asteroid['vy'], asteroid['vz']))
        })

# Create a DataFrame for better visualization
approaches_df = pd.DataFrame(closest_approaches)

# Sort the DataFrame by the closest distance
approaches_df = approaches_df.sort_values(by='Closest Distance (km)')

# Print the closest approaches
print("\nAsteroids with Closest Approaches to Earth (Ranked by Distance):")
print(approaches_df.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
import json

# Load data
with open('../data/asteroids_more_realistic.json', 'r') as file:
    data = json.load(file)

# Convert JSON data to DataFrame
df = pd.DataFrame(data)

# Extract position and velocity
df['x'] = df['position'].map(lambda d: d['x'])
df['y'] = df['position'].map(lambda d: d['y'])
df['z'] = df['position'].map(lambda d: d['z'])
df['vx'] = df['velocity'].map(lambda d: d['vx'])
df['vy'] = df['velocity'].map(lambda d: d['vy'])
df['vz'] = df['velocity'].map(lambda d: d['vz'])

# Constants
earth_x, earth_y, earth_z = 149600000.0, 0.0, 0.0
seconds_in_year = 365.25 * 24 * 3600
collision_threshold_km = 100000  # 50,000 km for collision risk

# Time points to visualize (in years)
time_points = [0, 100, 200, 300]

# Initialize list to store asteroids with collision risk
collision_risk_asteroids = []

fig = plt.figure(figsize=(15, 12))

for i, year in enumerate(time_points):
    ax = fig.add_subplot(2, 2, i + 1, projection='3d')
    # Calculate new positions
    x_new = df['x'] + df['vx'] * year * seconds_in_year / 1000
    y_new = df['y'] + df['vy'] * year * seconds_in_year / 1000
    z_new = df['z'] + df['vz'] * year * seconds_in_year / 1000

    # Distance to Earth
    distance_to_earth = np.sqrt((x_new - earth_x)**2 + (y_new - earth_y)**2 + (z_new - earth_z)**2)
    close_asteroids = distance_to_earth < collision_threshold_km

    # Add close asteroids to the list
    collision_ids = df.loc[close_asteroids, 'id'].tolist()
    collision_risk_asteroids.extend(collision_ids)

    # Plotting
    ax.scatter(x_new[~close_asteroids], y_new[~close_asteroids], z_new[~close_asteroids], color='black', s=1, alpha=0.5, label='Safe Asteroids')
    ax.scatter(x_new[close_asteroids], y_new[close_asteroids], z_new[close_asteroids], color='red', s=20, label='Collision Risk')
    ax.scatter([earth_x], [earth_y], [earth_z], color='blue', s=200, edgecolor='black', label='Earth')  # Make Earth more visible

    # Zoom in around Earth
    zoom_range = 150000  # Adjust to ensure Earth is visible
    ax.set_xlim(earth_x - zoom_range, earth_x + zoom_range)
    ax.set_ylim(earth_y - zoom_range, earth_y + zoom_range)
    ax.set_zlim(earth_z - zoom_range, earth_z + zoom_range)

    ax.set_xlabel('X (km)')
    ax.set_ylabel('Y (km)')
    ax.set_zlabel('Z (km)')
    ax.legend()
    ax.set_title(f"Asteroids relative to Earth at {year} years")

plt.tight_layout()
plt.show()

# Print the collision risk asteroids
print("Asteroids with collision risk:")
print(collision_risk_asteroids)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
import json

# Charger les données depuis le fichier JSON
with open('../data/asteroids.json', 'r') as f:
    asteroid_data = json.load(f)

# Convertir en DataFrame pour faciliter le traitement
df = pd.DataFrame(asteroid_data)

# Constantes pour les calculs
earth_x, earth_y, earth_z = 149600000.0, 0.0, 0.0
seconds_in_year = 365.25 * 24 * 3600
time_points = [0, 100, 200, 300]

fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111, projection='3d')

# Tracer les trajectoires des astéroïdes
for index, asteroid in df.iterrows():
    x_positions = [asteroid['position']['x'] + asteroid['velocity']['vx'] * year * seconds_in_year / 1000 for year in time_points]
    y_positions = [asteroid['position']['y'] + asteroid['velocity']['vy'] * year * seconds_in_year / 1000 for year in time_points]
    z_positions = [asteroid['position']['z'] + asteroid['velocity']['vz'] * year * seconds_in_year / 1000 for year in time_points]
    
    ax.plot(x_positions, y_positions, z_positions, label=asteroid['id'])

# Ajouter la Terre à la visualisation
ax.scatter([earth_x], [earth_y], [earth_z], color='blue', s=200, edgecolor='black', label='Earth')

# Ajuster les limites et les labels
ax.set_xlim(earth_x - 1e6, earth_x + 1e6)
ax.set_ylim(earth_y - 1e6, earth_y + 1e6)
ax.set_zlim(earth_z - 1e6, earth_z + 1e6)

ax.set_xlabel('X (km)')
ax.set_ylabel('Y (km)')
ax.set_zlabel('Z (km)')
ax.legend(loc='upper right')
ax.set_title("Trajectoires des astéroïdes")

plt.show()
```
